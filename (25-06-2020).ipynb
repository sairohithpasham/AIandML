```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
boston_df = pd.DataFrame(boston.data, columns=boston.feature_names)
```

```python
boston_df
```

```python
boston_df['MEDV']=boston.target
boston_df
```

```python
boston_df.corr()
```

```python
plt.figure(figsize=(14,14))
sns.heatmap(boston_df.corr(),annot=True,cmap="YlGnBu")
```

```python
sns.heatmap(boston_df.corr()[boston_df.corr()>=0.7],annot=True)
```

```python
correl = boston_df.corr()
cor = correl[correl>0.7]
sns.heatmap(cor,annot=True)
```

```python
cor = correl[(correl>0.7) & (correl<0.9)]
cor
```

```python
sns.heatmap(cor,annot=True)
```

```python
X = boston_df.drop(['MEDV'],axis=1)
y = boston_df[['MEDV']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
# Build the model on training data
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
## intercept and coeff values
print(lm.intercept_)
print(lm.coef_)
```

```python
c = pd.DataFrame([X_train.columns, lm.coef_]).T
c
```

```python
## checking the model performance on training data
```

```python
y_train_pred = lm.predict(X_train)
y_train_pred
```

```python
from sklearn.metrics import r2_score
print('R^2:',r2_score(y_train,y_train_pred))
print("adjusted R^2:", 1-((1-r2_score(y_train,y_train_pred))*(len(X_train)-1))/(len(X_train)-X_train.shape[1]-1))
```

```python
## regplot(y_train---> actual values and y_train_pred)
sns.regplot(y_train,y_train_pred)
```

```python
# visualize the difference between the actual values and predicted values
plt.scatter(y_train,y_train_pred)
plt.xlabel('Price')
plt.ylabel('Model predicted price')
```

```python
errors=y_train-y_train_pred
```

```python
plt.scatter(y_train_pred,errors)
plt.xlabel('predicted Price')
plt.ylabel('Errors')
```

```python
y_test_pred = lm.predict(X_test)
y_test_pred
```

```python
from sklearn.metrics import r2_score
print('R^2:',r2_score(y_test,y_test_pred))
print("adjusted R^2:", 1-((1-r2_score(y_test,y_test_pred))*(len(X_test)-1))/(len(X_test)-X_test.shape[1]-1))
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print('MAE:',mean_absolute_error(y_test,y_test_pred))
print('MSE:',mean_squared_error(y_test,y_test_pred))
print('RMSE:',np.sqrt(mean_squared_error(y_test,y_test_pred)))
```
