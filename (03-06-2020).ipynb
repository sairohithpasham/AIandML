```python
import time
list1=list(range(size))
list2=list(range(size))
start=time.time()
result=[x+y for x,y in zip(list1,list2)]
print("list:",time.time()-start)

arr1=np.arange(size)
arr2=np.arange(size)
start=time.time()
result=arr1+arr2
print("array:",time.time()-start)
```
